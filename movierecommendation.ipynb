## Tfidf Vectorizer.(To get unique words.)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

Tfid= TfidfVectorizer(stop_words ="english")
movie["Plot_summary"] = movie["Plot_summary"].fillna(" ")
matrix =Tfid.fit_transform(movie["Plot_summary"])
```

## Cosine-similarity (To find similarity between  movies)

```python
from sklearn.metrics.pairwise  import linear_kernel
cosine_similarity = linear_kernel(matrix,matrix)
cosine_similarity
```

## Indices (To get movie input as indices as well as output)

```python
indices  = pd.Series(movie.index , index  = movie["Title"]).drop_duplicates()
indices
```

```python
enumerate(cosine_similarity[1])
```

## Get Recommendation function.

```python
def get_recomend(title , cosine_similarity = cosine_similarity):
    pict = indices[title]
   # print("movie :",movie)
    sim_scores = list(enumerate(cosine_similarity[pict]))
    #print("sim_scores1 :",sim_scores)
    sim_scores = sorted(sim_scores,key = lambda x:x[1] ,reverse = True)
    #print("sim_scores2 :",sim_scores)
    sim_scores = sim_scores[1:11]
    #print("sim_scores3 :",sim_scores)
    movie_indices = [i[0] for i  in sim_scores]
    print("movie_indices:",movie_indices)
    return movie['Title'].iloc[movie_indices]
  
```
